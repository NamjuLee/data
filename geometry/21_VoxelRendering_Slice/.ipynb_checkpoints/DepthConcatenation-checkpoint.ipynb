```python
import datetime
import numpy as np

import sys, json, re, os, glob
import scipy.misc
import h5py
from PIL import Image, ImageMath
np.random.seed(0)
```

```python
class ImgCon():    
    def __init__(self, pathes):
        self.imgSequenceConcat = []
        self.Train = []
        self.folder = None
        self.imgPath = []
        self.imgSequence = []
        self.imgZD = []
        self.imgRendering = []
        self.imgSequenceConcat = []
        self.Train = []
        
        self.folder = pathes
        path = pathes + "/*"
        self.imgPath = [g for g in (glob.glob(path))]
        print(self.imgPath)
        self.GetImgsFromPath()
        self.GetImgZD()
        self.GetImgRendering()
        self.ProcessSequenceImgs()
        self.ProcessSquenceImgWithRendering()
        self.ProcessZDWithRendering()

    def GetImgZD(self):
        img = [i for i in self.imgPath if i[-6:] == "ZD.jpg"][0]
        self.imgZD = self.GetImgFromPath(img)
    def GetImgRendering(self):
        img = [i for i in self.imgPath if i[-7:] == "Img.jpg"][0]
        self.imgRendering = self.GetImgFromPath(img)
    def GetImgsFromPath(self, slice=16):
        for i in self.imgPath[:slice]:
            self.imgSequence.append( self.GetImgFromPath(i) )
        
    def GetImgFromPath(self, path):
        return Image.open(path)
        
    def ToString(self):
        pass

    def ProcessSequenceImgs(self):
        widths, heights = zip(*(i.size for i in self.imgSequence))
        total_width = 64*4 # sum(widths[:4])
        total_height = 65 * 4 #sum(heights[:4])
        intervalWidth = float(total_width / 4)
        intervalHeight= float(total_height / 4)
        new_im = Image.new('RGB', (total_width, total_height))
#         print(len(self.imgSequence[15]))
        count =0
        for i in range(4):
            for j in range(4):
                img = self.imgSequence[count]
                img = img.resize((65,65))
                new_im.paste(img, (int(intervalWidth * j),int(intervalHeight * i)))
                count+= 1
        self.imgSequenceConcat = new_im
    def ProcessSquenceImgWithRendering(self):
        theTotalExtendedWith = self.imgSequenceConcat.size[0] + self.imgRendering.size[0]
        new_im2 = Image.new('RGB', (theTotalExtendedWith, self.imgSequenceConcat.size[1]))
        new_im2.paste(self.imgSequenceConcat, (0,0))
        new_im2.paste(self.imgRendering, (self.imgSequenceConcat.size[0],0))
        self.Train.append(new_im2)
    def ProcessZDWithRendering(self):
        theTotalExtendedWith = self.imgZD.size[0] + self.imgRendering.size[0]
        new_im = Image.new('RGB', (theTotalExtendedWith, self.imgRendering.size[1]))
        new_im.paste(self.imgZD, (0,0))
        new_im.paste(self.imgRendering, (self.imgZD.size[0],0))
        self.Train.append(new_im)

    def SaveImg(self, path =""):
        if(len(path) > 0):
            name = self.folder       
            savePath = "./" + path + "/ " + name + "_a.jpg"
            self.Train[0].save(savePath)
            savePath = "./" + path + "/ " + name + "_b.jpg"
            self.Train[1].save(savePath)  
        else:
            name = self.folder       
            savePath = "./" + self.folder + "/ " + name + "_a.jpg"
            self.Train[0].save(savePath)
            savePath = "./" + self.folder + "/ " + name + "_b.jpg"
            self.Train[1].save(savePath)  

```
