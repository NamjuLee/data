```python
# Dataset Parameters
batch_size = 100
load_size = 280
fine_size = 256
c = 3

meanD = [0.5 for i in range(280)]
# data_mean = np.asarray([0.5,0.5,0.5])
# data_mean = np.asarray(meanD)
# data_mean = np.asarray([0.5])
data_mean = np.asarray([[[0.5,0.5,0.5]]])
```

```python

# Training Parameters
learning_rate = 0.001
dropout = 0.5 # Dropout, probability to keep units
training_iters = 500 #10000  # 50000 / batch_size # 5 ephoc  -- 8000 for 5hours
step_display = 10
step_save = 50
path_save = 'model/classificationChair' #  '/classification'
start_from =""#classification/classificationMomentumOptimizer-200"#./classification-10000" # "/classification-10000"#'/classification-10000.ckpt'
```

```python
# Loading data from disk
class DataLoaderDisk(object):
    def __init__(self, **kwargs):

        self.load_size = int(kwargs['load_size'])
        self.fine_size = int(kwargs['fine_size'])
        self.data_mean = np.array(kwargs['data_mean'])
        self.randomize = kwargs['randomize']
        self.data_root = os.path.join(kwargs['data_root'])

        # read data info from lists
        self.list_im = []
        self.list_lab = []
        with open(kwargs['data_list'], 'r') as f: # " PATH/box1.png 0"
            for line in f:
                path, lab = line.rstrip().split(' ')
                #self.list_im.append(os.path.join(self.data_root, path))
                self.list_im.append(path)
                self.list_lab.append(int(lab))
        self.list_im = np.array(self.list_im, np.object)
        self.list_lab = np.array(self.list_lab, np.int64)
        self.num = self.list_im.shape[0]
        print('# Images found:', self.num)

        # permutation
        perm = np.random.permutation(self.num) 
        self.list_im[:, ...] = self.list_im[perm, ...]
        self.list_lab[:] = self.list_lab[perm, ...]

        self._idx = 0
        
    def next_batch(self, batch_size):
        images_batch = np.zeros((batch_size, self.fine_size, self.fine_size, 3)) 
        labels_batch = np.zeros(batch_size)
        for i in range(batch_size):
            image = scipy.misc.imread(self.list_im[self._idx], mode="RGB")
            

            image = scipy.misc.imresize(image, (self.load_size, self.load_size), interp='nearest')

            
#             print(image)
#             limit = 220
#             for i in range(len(image)):
#                 for j in range(len(image[i])):
#                     if image[i][j][0] > limit and image[i][j][1] > limit and image[i][j][2] > limit :
#                         image[i][j][0] = 0
#                         image[i][j][1] = 0
#                         image[i][j][2] = 255
#                     elif image[i][j][0] > limit and image[i][j][1] < limit and image[i][j][2] < limit :
#                         image[i][j][0] = 255
#                         image[i][j][1] = 0
#                         image[i][j][2] = 0
#                     elif image[i][j][0] < limit and image[i][j][1] < limit and image[i][j][2] > limit :
#                         image[i][j][0] = 0
#                         image[i][j][1] = 255
#                         image[i][j][2] = 0
        
        
            image = image.astype(np.float32)/255.   
#             print(image.shape)
#             print("-------------------")
#             print(self.data_mean.shape)
#             print(self.data_mean)
#             print(image)      
#             image = image - self.data_mean
#             print(image)
#             imgplot = plt.imshow(image)
#             plt.show() 
#             break
            if self.randomize:
                flip = np.random.random_integers(0, 1)
                if flip>0:
                    image = image[:,::-1,:]
                offset_h = np.random.random_integers(0, self.load_size-self.fine_size)
                offset_w = np.random.random_integers(0, self.load_size-self.fine_size)
            else:
                offset_h = int((self.load_size-self.fine_size)/2)
                offset_w = int((self.load_size-self.fine_size)/2)

            images_batch[i, ...] =  image[offset_h:offset_h+self.fine_size, offset_w:offset_w+self.fine_size, :]
            labels_batch[i, ...] = self.list_lab[self._idx]
            
            self._idx += 1
            if self._idx == self.num:
                self._idx = 0
        
        return images_batch, labels_batch
    
    def size(self):
        return self.num

    def reset(self):
        self._idx = 0
```

```python
# Construct dataloader
opt_data_train = {
    #'data_h5': 'miniplaces_256_train.h5',
    'data_root': '.',   # MODIFY PATH ACCORDINGLY
    'data_list': 'trainAllRender.txt', # MODIFY PATH ACCORDINGLY
    'load_size': load_size,
    'fine_size': fine_size,
    'data_mean': data_mean,
    'randomize': False
    }
opt_data_val = {
    #'data_h5': 'miniplaces_256_val.h5',
    'data_root': '.',   # MODIFY PATH ACCORDINGLY
    'data_list': 'valAllRender.txt', #'val.txt',   # MODIFY PATH ACCORDINGLY
    'load_size': load_size,
    'fine_size': fine_size,
    'data_mean': data_mean,
    'randomize': True
    }

loader_train = DataLoaderDisk(**opt_data_train)
# print(loader_train)
loader_val = DataLoaderDisk(**opt_data_val)
#loader_train = DataLoaderH5(**opt_data_train)
#loader_val = DataLoaderH5(**opt_data_val)

loader_train.next_batch(100)
```

```python

def net(x, keep_dropout):
    weights = {
        'wc1': tf.Variable(tf.random_normal([11, 11, 3, 96], stddev=np.sqrt(2./(11*11*3)))),
        'wc2': tf.Variable(tf.random_normal([5, 5, 96, 256], stddev=np.sqrt(2./(5*5*96)))),
        'wc3': tf.Variable(tf.random_normal([3, 3, 256, 384], stddev=np.sqrt(2./(3*3*256)))),
#         'wc4': tf.Variable(tf.random_normal([3, 3, 384, 256], stddev=np.sqrt(2./(3*3*384)))),
#         'wc5': tf.Variable(tf.random_normal([3, 3, 256, 256], stddev=np.sqrt(2./(3*3*256)))),

        'wf6': tf.Variable(tf.random_normal([4*4*384 , 1024], stddev=np.sqrt(2./(4*4*256)))),
#        'wf7': tf.Variable(tf.random_normal([4096, 4096], stddev=np.sqrt(2./4096))),
        'wo': tf.Variable(tf.random_normal([1024, 13], stddev=np.sqrt(2./1024)))
    }

    biases = {
        'bc1': tf.Variable(tf.zeros(96)),
        'bc2': tf.Variable(tf.zeros(256)),
        'bc3': tf.Variable(tf.zeros(384)),
#         'bc4': tf.Variable(tf.zeros(256)),
#         'bc5': tf.Variable(tf.zeros(256)),

        'bf6': tf.Variable(tf.zeros(1024)),
#        'bf7': tf.Variable(tf.zeros(4096)),
        'bo': tf.Variable(tf.zeros(13))
    }

    # Conv + ReLU + LRN + Pool, 256->64->32
    conv1 = tf.nn.conv2d(x, weights['wc1'], strides=[1, 4, 4, 4, 1], padding='SAME')
    conv1 = tf.nn.relu(tf.nn.bias_add(conv1, biases['bc1']))
#     lrn1 = tf.nn.local_response_normalization(conv1, depth_radius=5, bias=1.0, alpha=1e-4, beta=0.75)
    pool1 = tf.nn.max_pool(conv1, ksize=[1, 3, 3,3, 1], strides=[1, 2, 2,2, 1], padding='SAME')

    # Conv + ReLU + LRN + Pool, 32-> 16 ->8
    conv2 = tf.nn.conv2d(pool1, weights['wc2'], strides=[1, 2, 2,2, 1], padding='SAME')
    conv2 = tf.nn.relu(tf.nn.bias_add(conv2, biases['bc2']))
#     lrn2 = tf.nn.local_response_normalization(conv2, depth_radius=5, bias=1.0, alpha=1e-4, beta=0.75)
    pool2 = tf.nn.max_pool(conv2, ksize=[1, 3, 3,3, 1], strides=[1, 2, 2, 1], padding='SAME')

    # Conv + ReLU, 8->4
    conv3 = tf.nn.conv2d(pool2, weights['wc3'], strides=[1,1, 1, 1], padding='SAME') # 1,1,1,1
    conv3 = tf.nn.relu(tf.nn.bias_add(conv3, biases['bc3']))
    
    pool3 = tf.nn.max_pool(conv3, ksize=[1, 3, 3, 1], strides=[1, 2, 2, 1], padding='SAME')

    # Conv + ReLU, 13-> 13
#     conv4 = tf.nn.conv2d(conv3, weights['wc4'], strides=[1, 1, 1, 1], padding='SAME')
#     conv4 = tf.nn.relu(tf.nn.bias_add(conv4, biases['bc4']))

    # Conv + ReLU + Pool, 13->6
#     conv5 = tf.nn.conv2d(conv4, weights['wc5'], strides=[1, 1, 1, 1], padding='SAME')
#     conv5 = tf.nn.relu(tf.nn.bias_add(conv5, biases['bc5']))
#     pool5 = tf.nn.max_pool(conv5, ksize=[1, 3, 3, 1], strides=[1, 2, 2, 1], padding='SAME')

    # FC + ReLU + Dropout
    fc6 = tf.reshape(pool3, [-1, weights['wf6'].get_shape().as_list()[0]])
    fc6 = tf.add(tf.matmul(fc6, weights['wf6']), biases['bf6'])
    fc6 = tf.nn.relu(fc6)
    fc6 = tf.nn.dropout(fc6, keep_dropout)
    
    # FC + ReLU + Dropout
#     fc7 = tf.add(tf.matmul(fc6, weights['wf7']), biases['bf7'])
#     fc7 = tf.nn.relu(fc7)
#     fc7 = tf.nn.dropout(fc7, keep_dropout)

    # Output FC
    out = tf.add(tf.matmul(fc6, weights['wo']), biases['bo'])
    
    return out
```

```python
# tf Graph input
x = tf.placeholder(tf.float32, [None, fine_size, fine_size, c])
y = tf.placeholder(tf.int64, None)
keep_dropout = tf.placeholder(tf.float32)

# Construct model
logits = net(x, keep_dropout)
print(logits)
print(y)
# labels=..., logits=
# Define loss and optimizer
loss = tf.reduce_mean(tf.nn.sparse_softmax_cross_entropy_with_logits(logits=logits, labels=y)) # logits=logits, labels=y

train_optimizer = tf.train.AdamOptimizer(learning_rate=learning_rate).minimize(loss)
# train_optimizer = tf.train.GradientDescentOptimizer(learning_rate=learning_rate).minimize(loss)
# train_optimizer = tf.train.MomentumOptimizer(learning_rate=learning_rate, momentum=0.9).minimize(loss)
# train_optimizer = tf.train.ProximalGradientDescentOptimizer(learning_rate=learning_rate).minimize(loss)
# train_optimizer = tf.train.FtrlOptimizer(learning_rate, learning_rate_power=-0.5).minimize(loss)
# train_optimizer = tf.train.RMSPropOptimizer(learning_rate, decay=0.9, momentum=0.0, epsilon=1e-10, use_locking=False, name='RMSProp').minimize(loss)

                       # Evaluate model
accuracy1 = tf.reduce_mean(tf.cast(tf.nn.in_top_k(logits, y, 1), tf.float32))
# accuracy5 = tf.reduce_mean(tf.cast(tf.nn.in_top_k(logits, y, 5), tf.float32))

# define initialization
init = tf.global_variables_initializer()

# define saver
saver = tf.train.Saver()

# define summary writer
#writer = tf.train.SummaryWriter('.', graph=tf.get_default_graph())
```

```python
dataResult = [];
# Launch the graph
with tf.Session() as sess:
    # Initialization
    if len(start_from)>1:
        saver.restore(sess, start_from)
        print("start from model...")
    else:
        sess.run(init)
        print("start from scratch...")
    step = 0

    while step < training_iters:
        # Load a batch of training data
        images_batch, labels_batch = loader_train.next_batch(batch_size)   
#         print("====================================")
#         print(images_batch.shape)
#         print(images_batch[0])
#         imgplot = plt.imshow(images_batch[0])
#         plt.show()
#         print("---------------")
#         print(labels_batch.shape)
#         print(labels_batch)
#         print("dd")
#         break


        if step % step_display == 0:
            print( '[%s]:' %(datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S"))) 
            ## Calculate batch loss and accuracy on training set
            l, acc1 = sess.run([loss, accuracy1], 
                               feed_dict={x: images_batch, y: labels_batch, keep_dropout: 1.}) 
            print( "-Iter " + str(step) + ", Training Loss= " + \
                   "{:.4f}".format(l) + ", Accuracy Top1 = " + \
                   "{:.2f}".format(acc1))
#             break
#             Calculate batch loss and accuracy on validation set
#             print("!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!")
#             print(batch_size)
            images_batch_val, labels_batch_val = loader_val.next_batch(batch_size) 
            log, l2, acc12 = sess.run([logits, loss, accuracy1], 
                                    feed_dict={x: images_batch_val, y: labels_batch_val, keep_dropout: 1.}) 
#             print("5555555555555555555555555555555555555555555555555555")
#             print(log)
            print ("-Iter " + str(step) + ", Validation Loss= " + \
                   "{:.4f}".format(l2) + ", Accuracy Top1 = " + \
                   "{:.2f}".format(acc12))
            
            dataResult.append([l,acc1, l2, acc12])
        # Run optimization op (backprop)
        sess.run(train_optimizer, feed_dict={x: images_batch, y: labels_batch, keep_dropout: dropout})
        
        step += 1
        
        
        # Save model
        if step % step_save == 0:
            saver.save(sess, path_save, global_step=step)
            print ("Model saved at Iter %d !" %(step))
        
    print("Optimization Finished!")
    saver.save(sess, path_save, global_step=9999)

    # Evaluate on the whole validation set
    print('Evaludation on the whole validation set...')
    num_batch = loader_val.size()/batch_size
#     print("00000000000000000000000000000000000000000000000")
#     print(loader_val.size())
#     print("00000000000000000000000000000000000000000000000")
    acc1_total = 0.
    acc5_total = 0.
    loader_val.reset()
    num_batch =int(num_batch)
    for i in range(num_batch):
        images_batch, labels_batch = loader_val.next_batch(batch_size)    
        [acc1] = sess.run([accuracy1], feed_dict={x: images_batch, y: labels_batch, keep_dropout: 1.})
        acc1_total += acc1
        print( "Validation Accuracy Top1 = " + \
               "{:.2f}".format(acc1))

    acc1_total /= num_batch
    print('Evaluation Finished! Accuracy Top1 = ' + "{:.4f}".format(acc1_total))
```
